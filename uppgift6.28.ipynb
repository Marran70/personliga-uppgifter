# skapa symboler

```python
from mtm026 import *
EI, L, x, Δ, q0 = sp.symbols("EI, L, x, Δ, q0")
```

```python
q = -q0
w = sp.Function("w")(x)

ode = sp.Eq(sp.diff(EI*sp.diff(w, x, 2), x, 2), q)
display(ode)
```

# integrara function

```python
w = sp.dsolve(ode, w).rhs
display(w)
```

```python
#randvillkor
wprim= w.diff(x)
wbis = w.diff(x, 2)

r1 = sp.Eq(w.subs(x, 0), 0)
r2 = sp.Eq(wprim.subs(x, 0), 0)
r3 = sp.Eq(w.subs(x, L/2 ), 0)
r4 = sp.Eq(wprim.subs(x, L/2), 0)

M = EI*wbis

display(r1, r2, r3, r4)
```

```python
#lösa integrationskonstanter
randvillkor = [r1, r2, r3, r4]
obekanta = ["C1", "C2", "C3", "C4"]
sol = sp.solve(randvillkor, obekanta)
displayvar("sol", sol)
```

```python
#ta fram momentekvation samt vanlig ekvation
w = w.subs(sol)

M = M.subs(sol)

display(M)
```
