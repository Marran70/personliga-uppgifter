```python
from mtm026 import *
E, I, L, P, Q = sp.symbols("E, I, L, P, Q")
```

Vi har 3 element alltså blir matrisen 8*8

```python
K = sp.zeros(8, 8)

K1 = Ke_balk(EI = E*I, L=L/2)
K2 = Ke_balk(EI=E*I, L = L/2)
K3 = Ke_balk(EI = E*I, L = L)
assem(K, K1, dofs=[1, 2, 3, 4])
assem(K, K2, dofs=[3, 4, 5, 6])
assem(K, K3, dofs=[5, 6, 7, 8])
display(K)
```

```python
#frihetsgrader
a1, a2, a3, a4, a5, a6, a7, a8 = sp.symbols("a1:9")
f1, f2, f3, f4, f5, f6, f7, f8 = sp.symbols("f1:9")
```

```python
a = sp.Matrix([0, a2, a3, a4, 0, a6, 0, a8])
fb = sp.Matrix([f1, 0, 0, 0, f5, 0, f7, 0])

f = sp.zeros(8, 1)
f13 = fe_balk(q = -Q/L, L = L)
assem(f, f13, dofs= [5, 6, 7, 8])
f[2] = -P
display(a, f)
```

```python
ekv_sys = sp.Eq(K*a, fb+f)
display(ekv_sys)
obekanta = [a2, a3, a4, a6, a8, f1, f5,f7]

sol = sp.solve(ekv_sys, obekanta)

a = a.subs(sol)
f = f.subs(sol)

display(a, f)
```
