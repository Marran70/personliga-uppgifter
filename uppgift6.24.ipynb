```python
from mtm026 import *
a1, a2, a3, a4, a5, a6, a7, a8, a9, a10, f1, f2, f9, f10, E,L, I, EI, P = sp.symbols("a1, a2, a3, a4, a5, a6, a7, a8, a9, a10 f1, f2, f9, f10 E, L, I, EI, P")
EI = E * I
k = EI/L**3

a = sp.Matrix([0, 0, a3, a4, a5, a6, a7, a8, 0, 0])
f = sp.Matrix([f1, f2, 0, 0, 0, 0, -P, 0, f9, f10])

Dofs = [
    [1, 2, 3, 4], 
    [5, 6, 7, 8], 
    [7, 8, 9, 10],
    [3, 5] #fjäderlement
]

K = sp.zeros(10, 10)

el_length = [L, L/2, L/2]

for el in range(3):
    dofs = Dofs[el]
    Ke = Ke_balk(EI, L = el_length[el])
    K = assem(K, Ke, dofs)

Kf = k * sp.Matrix([
    [1, -1],
    [-1, 1]
])

K = assem(K, Kf, Dofs[3])

obekanta = [a3, a4, a5, a6, a7, a8, f1, f2, f9, f10]
ekv_sys = sp.Eq(K*a, f)
sol = sp.solve(ekv_sys, obekanta)
display(sol)

delta1 = sol[a3]
delta2 = sol[a5]

F1 = k * delta1
F2 = k * delta2
F = F2 - F1
display(delta1, delta2, F)
```
