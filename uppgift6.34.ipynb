```python
from mtm026 import *
EI, L, x, Δ, q0 = sp.symbols("EI, L, x, Δ, q0")
```

```python
w = sp.Function("w")(x)
q = -q0/L
ode = sp.Eq(sp.diff(EI*sp.diff(w, x, 2), x, 2), q)
display(ode)
```

```python
w = sp.dsolve(ode, w).rhs
display((w))
```

```python
#randvillkor

wprim = w.diff(x)
wbis = w.diff(x, 2)

m = -EI*wbis

r1 = sp.Eq(w.subs(x, 0), 0)
r2 = sp.Eq(wprim.subs(x, 0), 0)
r3 = sp.Eq(w.subs(x, L), 0)
r4 = sp.Eq(m.subs(x, L), 0)

display(r1, r2, r3, r4)
```

```python
#lös integrations konstanter
ekvationer = [r1, r2, r3, r4]
obekanta = ["C1", "C2", "C3", "C4"]

sol = sp.solve(ekvationer, obekanta)

displayvar("sol", sol)

m = m.subs(sol)
w = w.subs(sol)
display(m)
display(w)

```
