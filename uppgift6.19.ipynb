# utbredd last över hela balken

```python
from mtm026 import *
E, I, L, q0, P = sp.symbols("E, I, L, q0, P")
```

```python
K = sp.zeros(8, 8)

K1 = Ke_balk(EI=E*I, L=L/3)
K2 = Ke_balk(EI=E*I, L=L/3)
K3 = Ke_balk(EI=E*I, L=L/3)

assem(K, K1, dofs=[1, 2, 3, 4])
assem(K, K2, dofs=[3, 4, 5, 6])
assem(K, K3, dofs=[5, 6, 7, 8])

display(K)
```

# fixa frihetsgrader

```python
a1, a2, a3, a4, a5, a6, a7, a8 = sp.symbols("a1:9")
r1, r2, r3, r4, r5, r6, r7, r8 = sp.symbols("r1:9")
```

# bestäm vektorerna med frihetsgrader samt reaktionskrafter

```python
a = sp.Matrix([0, a2, 0, a4, a5, a6, 0, a8])
r = sp.Matrix([r1, 0, r3, 0, -P, 0, r7, 0])
fy = sp.zeros(8, 1)
```

# lägg på den utbredda lasten som verkar över hela balken

```python
fu = fe_balk(-P/L, L = L/3) #denna behöver inte vara i någon loop i detta fall eftersom alla element är likadana
assem(fy, fu, dofs=[1, 2, 3, 4]) #ifall det skulle varit olika elementlängden eller flera utbredda laster
assem(fy, fu, dofs=[3, 4, 5, 6]) #så är det bättre att göra detta steget med en foor loop istället som assemblerar varje gång för varje element. 
assem(fy, fu, dofs=[5, 6, 7, 8])

display(fu)
```

```python
ekv_sys = sp.Eq(K*a, fy+r)
display(ekv_sys)
obekanta= [a2, a4, a5, a6, a8, r1, r3, r7]
sol = sp.solve(ekv_sys, obekanta)
a = a.subs(sol)
fy = fy.subs(sol)
display(a, fy)
```
