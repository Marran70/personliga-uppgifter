```python
from mtm026 import *
E, I , L, M0 = sp.symbols("E, I, L, M0")
```

```python
K = sp.zeros(6, 6) #pga 2 balkar
K1 = Ke_balk(EI = E*I, L=L)
K2 = Ke_balk(EI = E*I, L=L)

assem(K, K1, dofs=[1, 2, 3, 4])
assem(K, K2, dofs=[3, 4, 5, 6])


```

```python
a1, a2, a3, a4, a5, a6 = sp.symbols("a1:7")
f1, f2, f3, f4, f5, f6 = sp.symbols("f1:7")
a1 = a2 = a3 = a5 = 0
f4 = M0
f6 = 0
```

```python
a = sp.Matrix([a1, a2, a3, a4, a5, a6])
f = sp.Matrix([f1, f2, f3, f4, f5, f6])
obekanta = [a4, a6, f1, f2, f3, f5]
```

```python
ekv_sys = sp.Eq(K*a, f)
display(ekv_sys)
```

```python
sol = sp.solve(ekv_sys, obekanta)

a = a.subs(sol)
f = f.subs(sol)
display(f)
```
