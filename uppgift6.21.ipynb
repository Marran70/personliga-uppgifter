```python
from mtm026 import *
a1, a2, a3, a4, a5, a6, a7, f1, f2, f7, E,L, I, EI, P = sp.symbols("a1, a2, a3, a4, a5, a6, a7, f1, f2, f7 E, L, I, EI, P")
EI = E * I
k = (6*EI)/(L**3)

a = sp.Matrix([0, 0, a3, a4, a5, a6, 0])
f = sp.Matrix([f1, f2, 0, 0, -P, 0, f7])
K = sp.zeros(7, 7)
Dofs = [
    [1, 2, 3, 4],
    [3, 4, 5, 6],
    [7, 3]
]

for el in range(2):
    dofs = Dofs[el]
    Ke = Ke_balk(EI, L = L/2)
    K = assem(K, Ke, dofs)

Kf = k*sp.Matrix([
    [1, -1],
    [-1, 1]
])

K = assem(K, Kf, dofs=Dofs[2])

obekanta = [a3, a4, a5, a6, f1, f2, f7]

ekv_sys = sp.Eq(K*a, f)
sol = sp.solve(ekv_sys, obekanta)
delta = sol[a3]
F = k * delta
display(sol)
display(delta, F)
```
